```python
rows = []
for i, row in seattle_transcript_sep_2022_to_dec_2022.iterrows():
    transcript_df = pd.read_csv(row.transcript_as_csv_path)
    

    ## Using Spacy 
    for i in range(len(transcript_df)):
                if isinstance(transcript_df.text[i], float): continue
                doc = nlp(transcript_df.text[i])
                for entity in doc.ents:
                    rows.append(['Seattle',row.id, 'spacy', transcript_df.index[i],entity.text, entity.label_])

seattle_df = pd.DataFrame(rows, columns = ["city" , "session_id", "ner_model", "sentence_index", "named_entity_type", "entity"])
seattle_df
```

```python
ners = seattle_df
```

```python
seattle_rows = []
seattle_sentences_dfs = {}

for i, session in seattle_transcript_sep_2022_to_dec_2022.iterrows():
    
    seattle_sentences_df = pd.read_csv(session.transcript_as_csv_path)
    seattle_sentences_dfs[session.transcript_as_csv_path] = seattle_sentences_df.assign(session_id = session.id)
    
    for t in range(len(seattle_sentences_df)):
        doc = nlp(str(seattle_sentences_df.text[t]))
        for ent in doc.ents:
            if ent.label_ in ners:
                seattle_rows.append([session.id, seattle_sentences_df.index[t], ent.label_, ent.text, 'Seattle'])
                
seattle_df = pd.DataFrame(seattle_rows, columns=["session_id", "sentence_index", "named_entity_type", 
                                                 "entity", "city"])


seattle_all_sentence_df = pd.concat(seattle_sentences_dfs).reset_index()
seattle_all_sentence_df
```

```python
ner = oakland_df
oakland_rows = []
oakland_sentences_dfs = {}

for i, session in oakland_transcripts_sep_2022_to_dec_2022.iterrows():
    
    oakland_sentences_df = pd.read_csv(session.transcript_as_csv_path)
    oakland_sentences_dfs[session.transcript_as_csv_path] = oakland_sentences_df.assign(session_id = session.id)
    
    for t in range(len(oakland_sentences_df)):
        doc = nlp(str(oakland_sentences_df.text[t]))
        for entity in doc.ents:
            if ent.label_ in ner:
                oakland_rows.append([session.id, oakland_sentences_df.index[t], ent.label_, ent.text, 'Oakland'])
                
oakland_df = pd.DataFrame(oakland_rows, columns=["session_id", "sentence_index", "named_entity_type", 
                                                 "entity", "city"])


oakland_all_sentence_df = pd.concat(oakland_sentences_dfs).reset_index()
oakland_all_sentence_df
```

```python
ner = louisville_df
louisville_rows = []
louisville_sentences_dfs = {}

for i, session in louisville_transcript_sep_2022_to_dec_2022.iterrows():
    
    louisville_sentences_df = pd.read_csv(session.transcript_as_csv_path)
    louisville_sentences_dfs[session.transcript_as_csv_path] = louisville_sentences_df.assign(session_id = session.id)
    
    for t in range(len(louisville_sentences_df)):
        doc = nlp(str(louisville_sentences_df.text[t]))
        for entity in doc.ents:
            if ent.label_ in ner:
                louisville_rows.append([session.id, louisville_sentences_df.index[t], ent.label_, ent.text, 'Louisville'])
                
louisville_df = pd.DataFrame(louisville_rows, columns=["session_id", "sentence_index", "named_entity_type", 
                                                 "entity", "city"])


louisville_all_sentence_df = pd.concat(louisville_sentences_dfs).reset_index()
louisville_all_sentence_df
```

```python
all_sentences_df = pd.concat([louisville_all_sentence_df, oakland_all_sentence_df, seattle_all_sentence_df])
display(all_sentences_df)
```

```python
total_sentences_df = all_sentences_df.groupby(['session_id'])['index'].count().to_frame()
total_sentences_df = total_sentences_df.rename({"index": "total_sentences"}, axis=1)
```

```python
percentages = []
for i, row in all_df.iterrows():  
    # Divide index by total number of sentences
    percentages.append(row['sentence_index']/total_sentences_df.loc[row.session_id]['total_sentences'])


all_df["percentage"] = percentages 
all_df
```

```python
# Takes only rows that has Seattle as city

only_seattle_df = all_df[(all_df.city=="Seattle")]
only_seattle_df
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=20,20

g = sns.FacetGrid(only_seattle_df, row="ner_model", col="session_id", margin_titles=True)
g.map_dataframe(sns.stripplot, x="percentage")
_ = g.set_axis_labels("sentence_index", "recognized entities")
```

```python
import matplotlib.pyplot as plt
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("linear")

sns.boxplot(x="percentage", y="entity", data=only_seattle_df,
            whis=[0, 100], width=.6, palette="vlag")
sns.stripplot(x="percentage", y="entity", data=seattle_df,
              size=4, color=".3", linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
