```python
from cdp_data import CDPInstances, datasets
import pandas as pd
```

```python
# Four distinct annotation sets from: Seattle, Louisville, Oakland, Alameda

seattle_transcripts = datasets.get_session_dataset(
    CDPInstances.Seattle,  # specify the city (or county) council we want data from
    start_datetime="2020-01-01",  # YYYY-MM-DD format
    end_datetime="2023-04-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
)
seattle_transcripts
```

```python
louisville_transcripts = datasets.get_session_dataset(
    CDPInstances.Louisville,  # specify the city (or county) council we want data from
    start_datetime="2020-01-01",  # YYYY-MM-DD format
    end_datetime="2023-04-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
)
louisville_transcripts
```

```python
oakland_transcripts = datasets.get_session_dataset(
    CDPInstances.Oakland,  # specify the city (or county) council we want data from
    start_datetime="2020-01-01",  # YYYY-MM-DD format
    end_datetime="2023-04-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
    raise_on_error=False
)
oakland_transcripts
```

```python
alameda_transcripts = datasets.get_session_dataset(
    CDPInstances.Alameda,  # specify the city (or county) council we want data from
    start_datetime="2020-01-01",  # YYYY-MM-DD format
    end_datetime="2023-04-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
)
alameda_transcripts
```

```python
seattle_transcripts['muni'] = 'seattle'
louisville_transcripts['muni'] = 'louisville'
oakland_transcripts['muni'] = 'oakland'
alameda_transcripts['muni'] = 'alameda'
```

```python
# combine all four cities in a df

# Oakland not working

all_cities_df = pd.concat([seattle_transcripts, louisville_transcripts, oakland_transcripts, alameda_transcripts])
```

```python
all_cities_df
```

```python
# Randomly sample 50 events from all of the events from each council 

random_50_from_all_cities = all_cities_df.sample(n=50)
random_50_from_all_cities
```

```python
# get all sentences
sentences_session_df = {}

for i, session in random_50_from_all_cities.iterrows():
    
    sentence_df = pd.read_csv(session.transcript_as_csv_path)
    
    # to keep session_id information associated with each sentence, 
    # create another dataframe with session_id column and muni column
    sentences_session_df[session.transcript_as_csv_path] = sentence_df.assign(session_id = session.id, muni = session.muni)
    

sentences_session_df = pd.concat(sentences_session_df).reset_index()
sentences_session_df = sentences_session_df.drop(columns=['level_0', 'level_1'])

sentences_session_df # all sentence data for all 50 events (with session id)
```

```python
# Create chunks of 5 sentences at a time. 


# group by session_id
grouped = sentences_session_df.groupby('session_id')

chunks = []

for session_id, group in grouped:
    for i in range(0, len(group), 5):
        chunk = group.iloc[i:i+5]
        chunks.append(chunk)


cleaned_chunks = []

for df in chunks:
    df = df[['text', 'session_id', 'muni']]
    df = df.reset_index(drop=True)
    cleaned_chunks.append(df)
```

```python
chunks_list = []
session_id_list = []
muni_list = []

for chunk in cleaned_chunks:
    text = ''.join(str(chunk['text'].tolist())).replace('[','').replace(']','').replace("'",'').replace(',', ' ').replace('"', '')
    chunks_list.append(text)
    session_id_list.append(chunk['session_id'][0])
    muni_list.append(chunk['muni'][0])

chunks_df = pd.DataFrame({'text': chunks_list, 'session_id': session_id_list, 'muni': muni_list})
chunks_df
```

```python
# Randomly sample 1000 of the chunks

random_1000_from_chunks_df = chunks_df.sample(n=1000).reset_index()
random_1000_from_chunks_df.iloc[0].text
```

```python
random_1000_from_chunks_df
random_1000_df = random_1000_from_chunks_df.drop(columns=['index'])
random_1000_df
```

```python
# add a new column meta
random_1000_df['meta'] = random_1000_df.apply(lambda row: {'muni': row['muni'], 'session_id': row['session_id']}, axis=1)


cleaned_random_1000_df = random_1000_df.drop(columns=['muni', 'session_id'])

cleaned_random_1000_df
```

```python
cleaned_random_1000_df.to_json('annotation_set.jsonl', orient='records', lines=True)

```
