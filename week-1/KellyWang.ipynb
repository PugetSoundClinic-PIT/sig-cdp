```python
# Write the code for pulling data from a
# single month in a city of your choosing

# Be sure to store the transcript files locally
# And store them as a CSV

# Hint: The CDPInstances object has all the cities we support

# IMPORTANT: only pull a month of data, don't want this to take a
# long time to download or to fill up your hard drive

from cdp_data import CDPInstances, datasets

# Get a dataset of "city council sessions"
kingCounty_transcripts_aug_2022_to_sep_2022 = datasets.get_session_dataset(
    CDPInstances.KingCounty,  # specify the city (or county) council we want data from
    start_datetime="2022-08-01",  # YYYY-MM-DD format
    end_datetime="2022-09-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
)
kingCounty_transcripts_aug_2022_to_sep_2022
```

```python
# Find the mean, median, and max sentence duration in a random meeting
# (just pick a session to look at) from the data you have available

import pandas as pd
import statistics


example_session = kingCounty_transcripts_aug_2022_to_sep_2022.iloc[0]


example_transcript_df = pd.read_csv(example_session.transcript_as_csv_path)

sentence_duration_df = example_transcript_df.end_time - example_transcript_df.start_time


mean_sentence_duration_in_example_session = statistics.mean(sentence_duration_df)
median_sentence_duration_in_example_session = statistics.median(sentence_duration_df)
max_sentence_duration_in_example_session = max(sentence_duration_df)

print("mean sentence duration:", mean_sentence_duration_in_example_session)
print("median sentence duration:", median_sentence_duration_in_example_session)
print("max sentence duration:", max_sentence_duration_in_example_session)
```

```python
# OPTIONAL

# Get the mean, median, and max sentence durations for all sessions in the dataset
# you created and return a new dataframe with those stats
# i.e.
# session_id | mean | median | max
# ...        | ...  | ...    | ...


rows = []
for i in range(len(kingCounty_transcripts_aug_2022_to_sep_2022)):
    session = kingCounty_transcripts_aug_2022_to_sep_2022.iloc[i]
    transcript_df = pd.read_csv(session.transcript_as_csv_path)
    sentence_duration_df = transcript_df.end_time - transcript_df.start_time
    mean_sentence_duration = statistics.mean(sentence_duration_df)
    median_sentence_duration = statistics.median(sentence_duration_df)
    max_sentence_duration = max(sentence_duration_df)
    rows.append([session.id, mean_sentence_duration, median_sentence_duration, max_sentence_duration])

stats_df = pd.DataFrame(rows, columns=["session_id", "mean", "median", "max"])
stats_df
```
