```python
# Write the code for pulling data from a
# single month in a city of your choosing

# Be sure to store the transcript files locally
# And store them as a CSV

# Hint: The CDPInstances object has all the cities we support

# IMPORTANT: only pull a month of data, don't want this to take a
# long time to download or to fill up your hard drive
import cdp_data
import seaborn as sns

cdp_data.__version__, sns.__version__

from cdp_data import CDPInstances, datasets

seattle_transcripts_nov_2022_to_dec_2022 = datasets.get_session_dataset(
    CDPInstances.Seattle,  # specify the city (or county) council we want data from
    start_datetime="2022-11-01",  # YYYY-MM-DD format
    end_datetime="2022-12-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,  # store transcripts as CSVs for easy pandas reading
)
seattle_transcripts_nov_2022_to_dec_2022
```

```python
# Find the mean, median, and max sentence duration in a random meeting
# (just pick a session to look at) from the data you have available
example_session = seattle_transcripts_nov_2022_to_dec_2022.iloc[0]
example_session
```

```python
import pandas as pd

example_transcript_df = pd.read_csv(example_session.transcript_as_csv_path)
example_transcript_df
```

```python
# Find the mean, median, and max sentence duration in a random meeting
# (just pick a session to look at) from the data you have available

# lets look at the first three sentences
import statistics
import array as arr

meanAdd = 0
a = []
for _, row in example_transcript_df[:len(example_transcript_df)].iterrows():
    dif = (row.end_time - row.start_time)
    a.append(dif)
    meanAdd = meanAdd + dif
    
mean = meanAdd / len(example_transcript_df)

med = statistics.median(a)

max_sen = max(a)

print("mean", f"{mean}")
print("median", f"{med}")
print("max", f"{max_sen}")
    
```
