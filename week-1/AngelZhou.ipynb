```python
from cdp_data import CDPInstances, datasets

# Long Beach data in Oct to Nov 2022
long_beach_oct_nov_2022 = datasets.get_session_dataset(
    CDPInstances.LongBeach,  # specify the city (or county) council we want data from
    start_datetime="2022-10-01",  # YYYY-MM-DD format
    end_datetime="2022-11-01",  # YYYY-MM-DD format
    store_transcript=True,  # store transcripts locally for fast file reading
    store_transcript_as_csv=True,
)
```

```python
# Find the mean, median, and max sentence duration in a random meeting
# (just pick a session to look at) from the data you have available
```

```python
# second meeting
example_session = long_beach_oct_nov_2022.iloc[1]
example_session
```

```python
import pandas as pd

example_transcript_df = pd.read_csv(example_session.transcript_as_csv_path)
example_transcript_df
```

```python
import math

sentence_duration = example_transcript_df["end_time"]-example_transcript_df["start_time"]
mean_sd = sentence_duration.mean() 
median_sd = sentence_duration.median()
max_sd = sentence_duration.max()
print("mean:", mean_sd)
print("median:", median_sd)
print("max:", max_sd)
```

```python
# OPTIONAL

# Get the mean, median, and max sentence durations for all sessions in the dataset
# you created and return a new dataframe with those stats
# i.e.
# session_id | mean | median | max
# ...        | ...  | ...    | ...
```

```python
sd_df = pd.DataFrame(columns=["session_id", "mean", "median", "max"])
for session in long_beach_oct_nov_2022.iterrows():
    ex_session = session[1]
    session_id = ex_session.id
    example_transcript_df = pd.read_csv(ex_session.transcript_as_csv_path)
    sentence_duration = example_transcript_df["end_time"] - example_transcript_df["start_time"]
    temp_df = pd.DataFrame({'session_id': [session_id],
        'mean': [sentence_duration.mean()],
        'median': [sentence_duration.median()],
        'max': [sentence_duration.max()]})
    sd_df = pd.concat([sd_df, temp_df], ignore_index=True)
sd_df
```
